```python
import os
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv('golf.csv')
df
```

```python
x = df.iloc[:,:-1].values #x sebagai target
y = df.iloc[:,4].values #y sebagai prediktor
```

```python
# Membuat tabel Kontingensi
one_hot =pd.get_dummies(df[['Outlook','Temp.','Humidity','Wind']],drop_first=False,dtype=int)
one_hot.head()
```

```python
df = df.drop(['Outlook','Temp.','Humidity','Wind'],axis=1)
```

```python
x=pd.concat([one_hot,df],axis=1)
x.head()
# concat berguna untuk menggabungkan variabel
```

```python
# train and latih data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```
